```python
import numpy as np
import pandas as pd
import os
import json
import time
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from Policy_contextfree import *
from policy_LinUCB import *
from policy_HybridLinUCB import *


drop = 'https://www.petspyjamas.com/travel/nav/dog-friendly-cottages-hot-tub'
opt = 'https://www.petspyjamas.com/travel/hub/dog-friendly-cotswolds/'
def runepoch(policy,views,epoch = 1):
    mean_reward=[]
    cum_regret=[]
    mean_regret=[]
    for i in range(epoch):
        x,y,z = policy.policy_evaluator(views)
        mean_reward.append(x)
        cum_regret.append(y)
        mean_regret.append(z)
    mean_reward = np.average(mean_reward,axis=0)
    cum_regret = np.average(cum_regret,axis=0)
    mean_regret = np.average(mean_regret,axis=0)
    return mean_reward,cum_regret,mean_regret

#user_f = pd.read_csv("../../data/user_features/user_features.csv")
user_f = pd.read_csv("../../data/user_features/user_features_dummy.csv")
block_f = pd.read_csv('../../data/block_features/block_features.csv')
views = pd.read_csv('../../data/views.csv')
views = views[views.blockid!=drop].iloc[:3000]
#idx =  np.array(random.sample(range(len(views)), 5000))
#views = views.ix[idx]
#views = views.set_index(np.arange(len(views)))
block_f = block_f[block_f.blockid.isin(views.blockid.unique())]
blocks = list(block_f.blockid)
print(len(blocks))
pca = PCA(n_components=6)

pca_df = block_f[list(block_f.columns[1:])]
pca_df = pd.DataFrame(pca.fit_transform(pca_df))
pca_df.index = block_f.index
block_f = pd.concat([block_f[list(block_f.columns[:1])], pd.DataFrame(pca_df)] ,axis=1)

pca_df = user_f[list(user_f.columns[1:])]
pca_df = pd.DataFrame(pca.fit_transform(pca_df))
pca_df.index = user_f.index
user_f = pd.concat([user_f[list(user_f.columns[:1])], pd.DataFrame(pca_df)] ,axis=1)

```

```python
print('6')
linucb = LinUCB(0.1,user_f, block_f,1)
r_3,rr_3,rrr_3 = runepoch(linucb,views)
print('6')
hlinucb = LinUCB(0.3,user_f, block_f,1)
r_4,rr_4,rrr_4 = runepoch(hlinucb,views)
print('1')
e_greedy = LinUCB(0.5,user_f, block_f,1)
r_11,rr_11,rrr_11 = runepoch(e_greedy,views)
print('2')
e_greedy = LinUCB(0.7,user_f, block_f,1)
r_0,rr_0,rrr_0 = runepoch(e_greedy,views)
print('4')
ucb1 = LinUCB(0.9,user_f, block_f,1)
r_1,rr_1,rrr_1 = runepoch(ucb1,views)
print('5')
ts = LinUCB(1.2,user_f, block_f,1)
r_2,rr_2,rrr_2 = runepoch(ts,views)


import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
ax = plt.gca()
ax.plot(r_0, label='E_greedy 0.1')
ax.plot(r_11, label='E_greedy 1.0')
ax.plot(r_1, label='UCB1')
ax.plot(r_2, label='Thompson sampling')
ax.plot(r_3, label='LinUCB')
ax.plot(r_4, label='HybridLinUCB')

optimal=len(views[views['blockid']==opt])/len(views)
ax.axhline(y=optimal, color='k', linestyle=':', linewidth=2.5,label='optimal(hothub)')
ax.legend(loc='lower right')
plt.xlabel('T')
plt.ylabel('Cumulative average reward')
plt.show()

plt.figure(figsize=(12,8))
ax = plt.gca()
ax.plot(rr_0, label='E_greedy 0.1')
ax.plot(rr_11, label='E_greedy 1.0')
ax.plot(rr_1, label='UCB1')
ax.plot(rr_2, label='Thompson sampling')
ax.plot(rr_3, label='LinUCB')
ax.plot(rr_4, label='HybridLinUCB')

ax.legend(loc='upper left')
plt.xlabel('T')
plt.ylabel('Cumulative regret')
plt.show()

plt.figure(figsize=(12,8))
ax = plt.gca()
ax.plot(rrr_0, label='E_greedy 0.1')
ax.plot(rrr_11, label='E_greedy 1.0')
ax.plot(rrr_1, label='UCB1')
ax.plot(rrr_2, label='Thompson sampling')
ax.plot(rrr_3, label='LinUCB')
ax.plot(rrr_4, label='HybridLinUCB')

ax.legend(loc='upper right')
plt.xlabel('T')
plt.ylabel('Cumulative average regret')
plt.show()
```

```python
#user_f = pd.read_csv("../../data/user_features/user_features.csv")
user_f = pd.read_csv("../../data/user_features/user_features_dummy.csv")
block_f = pd.read_csv('../../data/block_features/block_features.csv')
views = pd.read_csv('../../data/views.csv')
block_f = block_f[block_f.blockid.isin(views.blockid.unique())]
blocks = list(block_f.blockid)

pca = PCA(n_components=6)

pca_df = block_f[list(block_f.columns[1:])]
pca_df = pd.DataFrame(pca.fit_transform(pca_df))
pca_df.index = block_f.index
block_f = pd.concat([block_f[list(block_f.columns[:1])], pd.DataFrame(pca_df)] ,axis=1)

pca_df = user_f[list(user_f.columns[1:])]
pca_df = pd.DataFrame(pca.fit_transform(pca_df))
pca_df.index = user_f.index
user_f = pd.concat([user_f[list(user_f.columns[:1])], pd.DataFrame(pca_df)] ,axis=1)
```

```python
linucb = LinUCB(1,user_f, block_f,1)
r=[]
file = ['06/22','06/23','06/24','06/25','06/26','06/27','06/28',
        '06/29','06/30','07/01','07/02','07/03','07/04','07/05',
        '07/06','07/07','07/08','07/09','07/10','07/11','07/12',
        '07/13','07/14','07/15','07/16','07/17','07/18','07/19',
        '07/20','07/21','07/22','07/23','07/24','07/25','07/26',
        '07/27','07/28','07/29','07/30','07/31']

np.save('views.npy',{})
views = load_dict('views.npy')
new = dict.fromkeys(blocks, 0)
for i in file:
    rewards = []
    logfile = "../../impression-2018/"+ i 
    g = os.walk(logfile)
    for path, d, filelist in sorted(g):
        for filename in sorted(filelist):
            file = open(os.path.join(path, filename))
            for line in file:
                l = json.loads(line)
                href = l.setdefault('meta').setdefault('href')
                if href in blocks:
                    userid = l.setdefault('cookie').split(';')[0].split('=')[1]
                    
                    if views == {}:
                        user_feature = np.zeros(len(blocks))
                        views[userid] = new.copy()
                    else:
                        if userid in views.keys():
                            user_feature = np.array(list(views[userid].values()))
                        else:
                            views[userid] = new.copy()
                            user_feature = np.zeros(len(blocks))
                    #user_feature = np.append(user_feature,user_f[user_f.userid == userid].iloc[:,1:].values[0])
                    
                    #user_feature = user_f[user_f.userid == userid].iloc[:,1:].values[0]
                    recommends = linucb.recommend(user_feature)
                    if href in recommends:
                        reward = 1
                        views[userid][href] = views[userid][href]+1
                    else:
                        reward = 0
                    linucb.update(reward)
                    rewards.append(reward)
                    
    avg_reward = np.average(np.array(rewards))
    r.append(avg_reward)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plt.plot(r)
plt.xlabel('Day')
plt.ylabel('average rewards')
plt.show()
```
