```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt, pymc3 as pm, theano
import seaborn as sns
%matplotlib inline
import datetime
```

```python
block_views = pd.read_csv('../data/views.csv')
blocks = pd.read_csv('../data/block_features/blocks.csv').rename(columns={'URL':'blockid'})
block_views = pd.merge(blocks,block_views)
```

```python
block_views['click'] = np.ones(len(block_views))
grouped = block_views.groupby(['userid', 'Collection '])[['click']].sum().reset_index()

frame=grouped.copy()
frame['UserID'] = pd.Categorical(grouped['userid'], categories=grouped['userid'].unique()).codes
frame['BlockID'] = pd.Categorical(grouped['Collection '], categories=grouped['Collection '].unique()).codes
#user_platform_unique=frame.groupby(['UserID','platform']).count()#user seems to use the same device
#user_ip_unique=frame.groupby(['UserID','REMOTE_ADDR']).count()#user seems to have different ips
```

```python
click_count = pd.DataFrame(frame.groupby('Collection ')['click'].count())
click_count = click_count.sort_values('click',ascending=False).reset_index()
plt.figure(figsize=(12,8))
sns.barplot(click_count['Collection '], click_count.click)
plt.ylabel("Frequency", fontsize=15)
# Add title
plt.xticks(rotation="vertical")
plt.xlabel("collection", fontsize=15)
plt.savefig('../datafig/block_popularity.png', dpi=96, bbox_inches='tight')
plt.show()
```

```python
user_click_count = pd.DataFrame(frame.groupby("UserID")["click"].count())
user_click_count = user_click_count.reset_index().click.value_counts()
plt.figure(figsize=(12,8))
sns.barplot(user_click_count.index, user_click_count.values)
plt.ylabel("Frequency", fontsize=15)
# Add title
plt.xticks(rotation="vertical")
plt.xlabel("Clicks per user", fontsize=15)
plt.savefig('../datafig/click_distribution.png', dpi=96, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x="date_week", data=frame)
plt.ylabel("Frquency")
plt.xlabel("Day of the week")
plt.title("Frequency distribution of clicks per day of the week")
plt.savefig('../datafig/week_distribution.png', dpi=96, bbox_inches='tight')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(12,8))
sns.countplot(x="date_hour", data=frame)
plt.ylabel("Frquency")
plt.xlabel("hour of the day")
plt.title("Frequency distribution of clicks per hour of the day")
plt.savefig('../datafig/day_distribution.png', dpi=96, bbox_inches='tight')
plt.show()
```
